```python
# Display the Nucleotides
ncov_dna
```

```python
# Length of our sequence
len(ncov_dna)
```

```python
# Transcribe (DNA to mRNA)
ncov_mRNA = ncov_dna.transcribe()
```

```python
# Changes Thymine to Uracil
ncov_mRNA
```

```python
# Translate to Protein/Amino Acids (mRNA to AA)
ncov_protein = ncov_mRNA.translate()
```

```python
ncov_protein
```

```python
# Length of Protein/Amino Acids
len(ncov_protein)
```

# **3D Structure of Covid**


*   File Format
  *   pdb :PDBParser() legacy
  *   cif :MMCIFParser() recent

***links***

* https://www.ncbi.nlm.nih.gov/Structure/pdb/6LU7  
* Protein Data Bank

```python
# Import Our Parser
from Bio.PDB import PDBParser,MMCIFParser
```

```python
# Read our pdb file
parser = PDBParser()
structure = parser.get_structure("6lu7","/content/drive/MyDrive/6lu7.pdb")
```

```python
structure
```

```python
# Chains in the Protein Structure
model = structure[0]
```

```python
for chain in model:
    print(f'chain {chain},chain_ID: {chain.id}')
```

```python
# Check the atoms
for model in structure:
    print(model)
    for chain in model:
        print(chain)
        for residue in chain:
            for atom in residue:
                print(atom)
```
